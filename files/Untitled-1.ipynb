```python
import pandas as pd

file_path = 'C:/Users/Hp/Desktop/Classeur1.csv'
data = pd.read_csv(file_path, encoding='latin1')

```

```python
# Remove the first row which is not part of the data
data = data.drop(index=0)

# Drop columns that contain all NaN values
data = data.dropna(axis=1, how='all')
data.head
```

```python
# Rename columns
data.columns = ['Alarm_ID', 'Alarm_name', 'Alarm_Sub_ID', 'Alarm_Cause', 'Severity', 'address_ip', 'bit']
```

```python
# Replace \n with space in all string columns
data = data.replace({r'\n': ' '}, regex=True)

# Handle missing values (if necessary)
data = data.fillna('')
```

```python
# Convert 'address_ip' and 'quantity' to integers
data['address_ip'] = data['address_ip'].astype(int)
data['quantity'] = data['quantity'].astype(int)

# # Convert 'gain' to integers, assuming missing values are replaced with 0
# data['gain'] = data['gain'].replace('', 0).astype(int)
```

```python
# Display the cleaned data
data
```

```python
data.to_csv('C:/Users/Hp/Desktop/cleaned_data.csv', index=False)
```
